### equation for CO2 plume:
        
$$c(x, y, z) = \cfrac{Qu}{2\pi\alpha_y\alpha_z u_*^2(x - x_s)^2}
\exp\left(-\cfrac{(y - y_s)^2u^2}{2\alpha_y^2u_*^2(x-x_s)^2} -\cfrac{(z - z_s)^2u^2}{2\alpha_z^2u_*^2(x-x_s)^2}\right)$$

````python
import numpy as np

Q = 0.26618286981003886
u = 0.4
pi = 3.14159265359
alpha_y = 0.066842568000323691
alpha_z = 0.14538827993452938
u_star = 0.06745668765535813
x_s = -0.64790143304753445

x_uw = -0.3
x_dw = 1

factor_uw = (Q*u) / (2*pi*alpha_y*alpha_z*(u_star**2)*(x_uw - x_s)**2)
std_y_uw = np.sqrt((alpha_y**2)*(u_star**2)*((x_uw - x_s)**2)/(u**2))
std_z_uw = np.sqrt((alpha_z**2)*(u_star**2)*((x_uw - x_s)**2)/(u**2))

print('uw factor', factor_uw)
print('uw std_y', std_y_uw)
print('uw std_z', std_z_uw)

factor_dw = (Q*u) / (2*pi*alpha_y*alpha_z*(u_star**2)*(x_dw - x_s)**2)
std_y_dw = np.sqrt((alpha_y**2)*(u_star**2)*((x_dw - x_s)**2)/(u**2))
std_z_dw = np.sqrt((alpha_z**2)*(u_star**2)*((x_dw - x_s)**2)/(u**2))

print('dw factor', factor_dw)
print('dw std_y', std_y_dw)
print('dw std_z', std_z_dw)
````
